### Prints
* `print,println,printf`

```java
public class Main {
    public static void main (String[] args){
        //normal print
        System.out.print("hello");
        //println
        System.out.println("hello");
        //normal print + \n --> same as println
        System.out.println("hello\n"); //will add 2 lines

        // Escape Characters \
        // to output one \ use : \\
        System.out.println("hello\\"); //outputs hello\
        // to jump one line use : \n
        System.out.println("hello\n");
        // to jump one tab use : \t
        System.out.println("\thello");

        //one line comment

        /*multiple
         * line 
         * comment
         * :) 
         */

        // SOUT + tab : to print System.out.print(); using intelliJ 
    }
}
Main mainInstance = new Main();
mainInstance.main(new String[]{});
```

### Variables
* variable: is a placeholder(marcador de posicin) for a value that behaves as the value it contains



```java
public class Main {
    public static void main (String[]args){
        // //declaration
        // int x;
        // //assigment
        // x = 25;
        // // declaration + assigment = initialization
        // int x = 25;

        //variables
        int a = 123;
        float b = 123.25f;
        double c = 123.25;
        boolean d = true; //or false
        char symbol = '@'; // in-between single coutes
        String name =  "Bruno";
    }
}
Main mainInstance = new Main();
mainInstance.main(new String[]{});
```
