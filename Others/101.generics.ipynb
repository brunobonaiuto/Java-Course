### Generics
* Enable types( which are classes and interfaces) to be parameters when defining classes, interfaces and methods
* a benefit is that: to eliminate the need to create multiple versions of methods or classes for variours data types.
* so basically we can use 1 version for all reference data types.

```java
public class Main{
    public static void main(String[]args){
        //we have four different arrays
        //these are reference data types
        Integer[] intArray = {1,2,3,4,5};
        Double[] doubleArray = {5.5, 4.4, 3.3, 2.2, 1.1};
        Character[] charArray = {'H','E','L','L','O'};
        String[] stringArray = {"B","Y","E"};
    }
}
Main.main(new String[]{});
```

* How do we display all the elements in the array?
    * we need a method for each data type

```java
public class Main{
    public static void main(String[]args){
        //we have four different arrays
        //these are reference data types
        Integer[] intArray = {1,2,3,4,5};
        Double[] doubleArray = {5.5, 4.4, 3.3, 2.2, 1.1};
        Character[] charArray = {'H','E','L','L','O'};
        String[] stringArray = {"B","Y","E"};

        displayArray(intArray);
        displayArray(doubleArray);
        displayArray(charArray);
        displayArray(stringArray);
    }
    //method to display the array of Integer    
    public static void displayArray(Integer[] arrayArg){
        //for each loop
        for (Integer x : arrayArg){
            System.out.print(x + ", ");
        }
        System.out.println();
    }
    //method to display the array of Double
    public static void displayArray(Double[] arrayArg){
        //for each loop
        for (Double x : arrayArg){
            System.out.print(x + ", ");
        }
        System.out.println();
    }
    //method to display the array of Char
    public static void displayArray(Character[] arrayArg){
        //for each loop
        for (Character x : arrayArg){
            System.out.print(x + ", ");
        }
        System.out.println();
    }
    //method to display the array of String
    public static void displayArray(String[] arrayArg){
        //for each loop
        for (String x : arrayArg){
            System.out.print(x + ", ");
        }
        System.out.println();
    }

}
Main.main(new String[]{});
```

* Using generics, we need just one method
    * just antes del return type, agrega <>
    * la gente normalmente agrega una T , ex: `<T>`
    * en lugar de la T, usare Thing
        * por que recibimos una cosa, retornamos otra cosa, e iteramos en cosas
* this example does not return nothing

```java
public class Main{
    public static void main(String[]args){
        //we have four different arrays
        //these are reference data types
        Integer[] intArray = {1,2,3,4,5};
        Double[] doubleArray = {5.5, 4.4, 3.3, 2.2, 1.1};
        Character[] charArray = {'H','E','L','L','O'};
        String[] stringArray = {"B","Y","E"};

        displayArray(intArray);
        displayArray(doubleArray);
        displayArray(charArray);
        displayArray(stringArray);
    }
    //method to display all types of arrays
    public static <Thing> void displayArray(Thing[] arrayArg){
        //for each loop
        for (Thing x : arrayArg){
            System.out.print(x + ", ");
        }
        System.out.println();
    }
}
    Main.main(new String[]{});
```

* To avoid creating one class for each data type we use a generic class

```java
public class Main{
    public static void main(String[] args){
        MyGenericClass myInt = new MyGenericClass(25);
        MyGenericClass myDouble = new MyGenericClass(25.5);
        MyGenericClass myChar = new MyGenericClass('B');
        MyGenericClass myString = new MyGenericClass("Bruno");

        System.out.println(myInt.getValue());
        System.out.println(myDouble.getValue());
        System.out.println(myChar.getValue());
        System.out.println(myString.getValue());
        
    }
}
//each of the classes should have
public class MyGenericClass <Thing> {
    //field
    Thing x; //store the thing variable
    //constructor
    MyGenericClass(Thing x){ //accept a thing called x
        this.x = x;
    }
    //method to return the value
    public Thing getValue(){ //return a thing
        return x;
    }
}
//...in the same way shoudl be for: MyDoubleClass, MyCharacterClass and MyStringClass
Main.main(new String[]{});
```

* pero hay una cosa mas por corregir incluso si funciona
* add the the `<T>` after the class name in the creation of the object
* and also add `< >` after the class name in construction of object

```java
public class Main{
    public static void main(String[] args){
        MyGenericClass<Integer> myInt = new MyGenericClass<>(25);
        MyGenericClass<Double> myDouble = new MyGenericClass<>(25.5);
        MyGenericClass<Character> myChar = new MyGenericClass<>('B');
        MyGenericClass<String> myString = new MyGenericClass<>("Bruno");

        System.out.println(myInt.getValue());
        System.out.println(myDouble.getValue());
        System.out.println(myChar.getValue());
        System.out.println(myString.getValue());
        
    }
}
//each of the classes should have
public class MyGenericClass <Thing> {
    //field
    Thing x; //store the thing variable
    //constructor
    MyGenericClass(Thing x){ //accept a thing called x
        this.x = x;
    }
    //method to return the value
    public Thing getValue(){ //return a thing
        return x;
    }
}
//...in the same way shoudl be for: MyDoubleClass, MyCharacterClass and MyStringClass
Main.main(new String[]{});
```

* there can be also multiples types, the common convention is `<T, V>`
    * we gonna call v : Thing2

```java
public class Main {
    public static void main(String[] args) {
        MyGenericClass<Integer,Integer> myInt = new MyGenericClass<>(1,1998);
        MyGenericClass<Double,Double> myDouble = new MyGenericClass<>(25.5,1998.0);
        MyGenericClass<Character,Character> myChar = new MyGenericClass<>('B','M');
        //mixed
        MyGenericClass<String,Character> myString = new MyGenericClass<>("Bruno",'M');

        System.out.println(myInt.getValue());
        System.out.println(myDouble.getValue());
        System.out.println(myChar.getValue());
        System.out.println(myString.getValue());

    }
}

//each of the classes should have
public class MyGenericClass<Thing1, Thing2> {
    //Fields
    Thing1 x;
    Thing2 y;
    //constructor
    MyGenericClass(Thing1 x, Thing2 y){
        this.x = x;
        this.y = y;
    }
    //get only thing2
    public Thing2 getValue(){
        return y;
    }
}
//...in the same way shoudl be for: MyDoubleClass, MyCharacterClass and MyStringClass
Main.main(new String[]{});
```
