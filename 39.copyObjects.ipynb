### Copy Objects in Java
* Wrong way of copying the content of obj1 into obj2

* Doing `obj1 = obj2` is wrong
    * This assing the same address for the two objects

```java
public class Car {
    
    private String make;
    private String model;
    private int year;

    Car(String make, String model, int year){
        //this.make = make;
        this.setMake(make);
        this.model = model;
        this.year = year;
    }

    public void setMake(String make){
        this.make = make;
    }
}

public class Main {
    public static void main(String[]args){
        //Copy the values of car1 into car2
        Car car1 = new Car("Chevrolet","Camaro", 2021);
        Car car2 = new Car("Ford","Mustang", 2023);
    
        //Wrong copy
        car2 = car1;
    }
}

Main.main(new String[]{});
```

* The right way is to create a method copy

```java
public class Car {
    
    private String make;
    private String model;
    private int year;

    Car(String make, String model, int year){
        //this.make = make;
        this.setMake(make);
        this.model = model;
        this.year = year;
    }
    //Getters
    public String getMake(){
        return make;
    }
    public String getModel(){
        return model;
    }
    public int getYear(){
        return year;
    }
    //Setters
    public void setMake(String make){
        this.make = make;
    }
    public void setModel(String make){
        this.model = model;
    }
    public void setYear(int make){
        this.year = year;
    }
    //Copy method
    public void copy(Car o){
        this.setMake(o.getMake());
        this.setModel(o.getModel());
        this.setYear(o.getYear());
    }
}

public class Main {
    public static void main(String[]args){
        //Copy the values of car1 into car2
        Car car1 = new Car("Chevrolet","Camaro", 2021);
        Car car2 = new Car("Ford","Mustang", 2023);
    
        //Right copy
        car2.copy(car1);      
        
        //Addresses should be different, and each object should have the same attributes
    }
}

Main.main(new String[]{});
```

```java
public class Car {
    
    private String make;
    private String model;
    private int year;

    Car(String make, String model, int year){
        //this.make = make;
        this.setMake(make);
        this.model = model;
        this.year = year;
    }
    //Copy constructor
    Car(Car carToCopy){
        this.copy(carToCopy);
    }

    //Getters
    public String getMake(){
        return make;
    }
    public String getModel(){
        return model;
    }
    public int getYear(){
        return year;
    }
    //Setters
    public void setMake(String make){
        this.make = make;
    }
    public void setModel(String make){
        this.model = model;
    }
    public void setYear(int make){
        this.year = year;
    }
    //Copy method
    public void copy(Car o){
        this.setMake(o.getMake());
        this.setModel(o.getModel());
        this.setYear(o.getYear());
    }
}

public class Main {
    public static void main(String[]args){
        //Copy the values of car1 into car2
        Car car1 = new Car("Chevrolet","Camaro", 2021);
        //Copying from constructor
        Car car2 = new Car(car1);
    
        //Addresses should be different, and each object should have the same attributes
    }
}

Main.main(new String[]{});
```
